```python
import pandas as pd
import numpy as np
from sklearn.linear_model import ElasticNetCV
from sklearn.preprocessing import StandardScaler

# Load the filtered data
df_filter = pd.read_csv('FOR_FILTER_FINAL.csv')

# Prepare training data (Scale features for penalty-based models)
X = df_filter.drop(columns=['id', 'los'])
y = df_filter['los']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- RUN ELASTIC NET CROSS-VALIDATION ---
# l1_ratio=0.5 provides an equal balance of Lasso and Ridge regularization
en_cv = ElasticNetCV(l1_ratio=0.5, cv=5, random_state=42, max_iter=2000)
en_cv.fit(X_scaled, y)

# Identify selected features (non-zero coefficients)
selected_features = X.columns[np.where(en_cv.coef_ != 0)[0]].tolist()
final_columns = ['id', 'los'] + selected_features

# Save final files ready for Machine Learning (XGBR/RFR)
df_ml = df_filter[final_columns]
df_ml.to_csv('FOR_ML.csv', index=False)

print(f"Final Feature Count for ML: {len(selected_features)}")
```
