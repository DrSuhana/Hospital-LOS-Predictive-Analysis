```python
import pandas as pd

df = pd.read_csv('FOR_FILTER_NUMERICAL.csv')

# Protect target and identifier
protected_cols = ['id', 'los']
feature_cols = [c for c in df.columns if c not in protected_cols]

# Calculate 1% Threshold for inclusion
threshold = 0.01 * len(df)

# Determine which features to retain (e.g., common ICD code blocks)
features_to_keep = [col for col in feature_cols if df[col].sum() >= threshold]

# Create filtered dataset
df_filtered = df[protected_cols + features_to_keep]

print(f"Features Retained: {len(features_to_keep)} (Dropped {len(feature_cols) - len(features_to_keep)} sparse features)")
df_filtered.to_csv('FOR_FILTER_FINAL.csv', index=False)
```
